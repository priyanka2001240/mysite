```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/heart.csv')
data=data.dropna(how='any',inplace=False)
print(data)
```

```python
sns.heatmap(data.isnull())
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

using KNN algorithm for finding accurracy

```python
from sklearn.neighbors import KNeighborsClassifier
knmodel=KNeighborsClassifier(n_neighbors=7)
knmodel.fit(x_train,y_train)
```

```python
y_pred=knmodel.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cn=confusion_matrix(y_test,y_pred)
print(cn)
sns.heatmap(cn,annot=True)
```

```python
 ac=accuracy_score(y_test,y_pred)
 print(ac)
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```

logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logm=LogisticRegression()
logm.fit(x_train,y_train)
```

```python
logm_pred=logm.predict(x_test)
logmcm=confusion_matrix(y_test,logm_pred)
# print(logmcm)
sns.heatmap(logmcm,annot=True)
logmac=accuracy_score(y_test,logm_pred)
print(logmac)
logmcr=classification_report(y_test,logm_pred)
print(logmcr)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
nb_pred=nb.predict(x_test)
nbcm=confusion_matrix(y_test,logm_pred)
# NeighborhoodComponentsAnalysis=confusion_matrix(y_test,nb_pred)
# print(logmcm)
sns.heatmap(nbcm,annot=True)
nbac=accuracy_score(y_test,nb_pred)
print(nbac)
nbcr=classification_report(y_test,nb_pred)
print(nbcr)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt_pred=dt.predict(x_test)
dtcm=confusion_matrix(y_test,dt_pred)
sns.heatmap(dtcm,annot=True)
dt_ac=accuracy_score(y_test,dt_pred)
print(dt_ac)
dt_cr=classification_report(y_test,dt_pred)
print(dt_cr)
```

```python
dt_pred=dt.predict(x_test)

dtcm=confusion_matrix(y_test,dt_pred)
# print(dtcm)
sns.heatmap(dtcm,annot=True)

dtac=accuracy_score(y_test,dt_pred)
print(dtac)

dtcr=classification_report(y_test,dt_pred)
print(dtcr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)

rfcm=confusion_matrix(y_test,rf_pred)
# print(dtcm)
sns.heatmap(rfcm,annot=True)

rfac=accuracy_score(y_test,rf_pred)
print(rfac)

rfcr=classification_report(y_test,rf_pred)
print(rfcr)
```

```python
joblib.dump(rf,"heartrf.joblib")
```
